```python
closes = list()
closes.append(np.random.uniform(60, 65, 10))
closes.append(np.random.uniform(65, 70, 10))
closes.append(np.random.uniform(70, 75, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(70, 75, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(90, 95, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(70, 75, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(90, 95, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(65, 70, 10))
closes.append(np.random.uniform(70, 75, 10))
closes.append(np.random.uniform(60, 65, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(90, 95, 10))
```

```python
close = np.array([])
for item in closes:
    close = np.append(close, item)
```

```python
values = talib.DEMA(close, timeperiod=20)
```

```python
seconds = range(0, len(close))
df = pd.DataFrame(zip(seconds, close), columns=['Time in seconds', 'Price'])
```

```python
df
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x = 'Time in seconds', y = 'Price', data=df)
```

## Adding values of Double Exponential Moving average

```python
df['DEMA'] = values
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x = 'Time in seconds', y = 'Price', data=df)
sns.lineplot(x = 'Time in seconds', y = 'DEMA', data=df)
```
