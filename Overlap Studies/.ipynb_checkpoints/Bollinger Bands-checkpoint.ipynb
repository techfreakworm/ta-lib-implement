```python
import numpy as np
import talib
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
closes = list()
closes.append(np.random.uniform(60, 65, 10))
closes.append(np.random.uniform(65, 70, 10))
closes.append(np.random.uniform(70, 75, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(70, 75, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(90, 95, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(70, 75, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(90, 95, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(80, 85, 10))
closes.append(np.random.uniform(65, 70, 10))
closes.append(np.random.uniform(70, 75, 10))
closes.append(np.random.uniform(60, 65, 10))
closes.append(np.random.uniform(75, 80, 10))
closes.append(np.random.uniform(90, 95, 10))
```

```python
close = np.array([])
for item in closes:
    close = np.append(close, item)
```

```python
close
```

```python
upper, middle, lower = talib.BBANDS(close, timeperiod=20)
```

```python
df = pd.DataFrame()
```

```python
df['points'] = close
```

```python
df = df.reset_index()
df.columns = ['Time in Seconds', 'Price']
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x = 'Time in Seconds', y = 'Price', data=df)
```

```python
df['upperband'] = upper
df['middleband'] = middle
df['lowerband'] = lower
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x = 'Time in Seconds', y = 'Price', data=df, label='Price')
sns.lineplot(x = 'Time in Seconds', y = 'upperband', data=df, label='Upper Band')
sns.lineplot(x = 'Time in Seconds', y = 'middleband', data=df, label='Middle Band')
ax = sns.lineplot(x = 'Time in Seconds', y = 'lowerband', data=df, label='Lower Band')
ax.set(ylabel='Price')
```
